```python
# =========================================================
# AI SMART CANTEEN - FINAL FIXED VERSION
# WEBCAM = TOA DO CUNG
# UPLOAD = AUTO SCALE
# =========================================================

import os
import cv2
import numpy as np
import tensorflow as tf
import tf_keras
import matplotlib.pyplot as plt
import gradio as gr

from tf_keras.models import load_model

from PIL import Image, ImageOps

from reportlab.lib.pagesizes import letter
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    Table,
    TableStyle
)

from reportlab.lib.styles import (
    getSampleStyleSheet,
    ParagraphStyle
)

from reportlab.lib import colors

# =========================================================
# INFO
# =========================================================

print("TensorFlow:", tf.__version__)

# =========================================================
# LOAD MODEL
# =========================================================

np.set_printoptions(suppress=True)

MODEL_PATH = "/content/ainhandiendoan.h5"

MENU_PATH = "/content/menu.txt"

try:

    model = load_model(
        MODEL_PATH,
        compile=False
    )

    class_names = [
        line.strip()
        for line in open(
            MENU_PATH,
            "r",
            encoding="utf-8"
        ).readlines()
    ]

    print("MODEL LOAD SUCCESS")

except Exception as e:

    print("MODEL LOAD FAILED")
    print(e)

    model = None

    class_names = []

# =========================================================
# CONFIG
# =========================================================

QR_PATH = "/content/QR.jpg"

PDF_OUTPUT_PATH = "hoa_don_khay_com.pdf"

FONT_NAME = "Helvetica"

# =========================================================
# PRICE MAP
# =========================================================

PRICE_MAP = {

    "ca": {
        "name": "Ca hu",
        "price": 30000
    },

    "canh rau": {
        "name": "Canh rau",
        "price": 7000
    },

    "canh chua": {
        "name": "Canh chua",
        "price": 10000
    },

    "canh chua co ca": {
        "name": "Canh chua co ca",
        "price": 25000
    },

    "com": {
        "name": "Com",
        "price": 10000
    },

    "dau hu": {
        "name": "Dau hu sot ca",
        "price": 25000
    },

    "rau": {
        "name": "Rau xao",
        "price": 10000
    },

    "suon": {
        "name": "Suon",
        "price": 30000
    },

    "thit kho": {
        "name": "Thit kho",
        "price": 25000
    },

    "thit kho trung": {
        "name": "Thit kho trung",
        "price": 30000
    },

    "trung": {
        "name": "Trung chien",
        "price": 25000
    }
}

# =========================================================
# PDF
# =========================================================

def generate_pdf_invoice(items, total_amount):

    doc = SimpleDocTemplate(
        PDF_OUTPUT_PATH,
        pagesize=letter
    )

    story = []

    styles = getSampleStyleSheet()

    title_style = ParagraphStyle(
        'Title',
        parent=styles['Heading1'],
        fontName=FONT_NAME,
        fontSize=20,
        alignment=1
    )

    story.append(
        Paragraph(
            "HOA DON THANH TOAN",
            title_style
        )
    )

    story.append(Spacer(1,20))

    table_data = [[
        "Vi tri",
        "Mon",
        "AI",
        "Gia"
    ]]

    for item in items:

        table_data.append([

            item["slot"],

            item["name"],

            f"{item['confidence']:.1f}%",

            f"{item['price']:,.0f} VND"
        ])

    table_data.append([
        "",
        "",
        "TONG",
        f"{total_amount:,.0f} VND"
    ])

    table = Table(table_data)

    table.setStyle(TableStyle([

        ('BACKGROUND',(0,0),(-1,0),
         colors.lightgrey),

        ('GRID',(0,0),(-1,-1),
         1,
         colors.black),

        ('FONTNAME',(0,0),(-1,0),
         'Helvetica-Bold')
    ]))

    story.append(table)

    doc.build(story)

    return PDF_OUTPUT_PATH

# =========================================================
# PROCESS
# =========================================================

def process_and_bill(input_img, tray_layout):

    if input_img is None:

        return (
            None,
            "<h3>Vui long chup anh!</h3>",
            None,
            None
        )

    if model is None:

        return (
            None,
            "<h3>Model chua load duoc!</h3>",
            None,
            None
        )

    # =====================================================
    # IMAGE
    # =====================================================

    original_image = Image.fromarray(
        input_img
    ).convert("RGB")

    img_w, img_h = original_image.size

    print("IMAGE:", img_w, img_h)

    # =====================================================
    # BOXES
    # =====================================================

    if tray_layout == "3 TOP - 2 BOTTOM":

        boxes_ratio = [

            (0.05, 0.03, 0.30, 0.42),

            (0.35, 0.03, 0.65, 0.42),

            (0.68, 0.03, 0.95, 0.42),

            (0.04, 0.45, 0.38, 0.92),

            (0.52, 0.45, 0.96, 0.92)
        ]

    else:

        boxes_ratio = [

            (0.03, 0.03, 0.47, 0.42),

            (0.53, 0.03, 0.97, 0.42),

            (0.03, 0.45, 0.32, 0.92),

            (0.35, 0.45, 0.65, 0.92),

            (0.68, 0.45, 0.97, 0.92)
        ]

    boxes_config = []

    for (rx1, ry1, rx2, ry2) in boxes_ratio:

        x1 = int(rx1 * img_w)
        y1 = int(ry1 * img_h)

        x2 = int(rx2 * img_w)
        y2 = int(ry2 * img_h)

        boxes_config.append(
            (x1, y1, x2, y2)
        )

    # =====================================================
    # DEBUG IMAGE
    # =====================================================

    debug_img = np.array(original_image).copy()

    for (x1, y1, x2, y2) in boxes_config:

        cv2.rectangle(
            debug_img,
            (x1, y1),
            (x2, y2),
            (0,255,0),
            5
        )

    # =====================================================
    # HTML
    # =====================================================

    html = """

    <div style='
        background:#0f172a;
        padding:25px;
        border-radius:20px;
        color:white;
        font-family:Segoe UI;
    '>

    <h2 style='color:#a855f7;text-align:center;'>
        HOA DON THANH TOAN
    </h2>

    """

    total_bill = 0

    pdf_items = []

    # =====================================================
    # PLOT
    # =====================================================

    fig, axes = plt.subplots(
        1,
        5,
        figsize=(18,4)
    )

    # =====================================================
    # LOOP
    # =====================================================

    for i, (left, upper, right, lower) in enumerate(boxes_config):

        crop = original_image.crop(
            (left, upper, right, lower)
        )

        crop_np = np.array(crop)

        crop_np = cv2.convertScaleAbs(
            crop_np,
            alpha=1.2,
            beta=10
        )

        crop = Image.fromarray(crop_np)

        processed = ImageOps.fit(
            crop,
            (224,224),
            Image.Resampling.LANCZOS
        )

        image_array = np.asarray(processed)

        normalized = (
            image_array.astype(np.float32) / 127.5
        ) - 1

        data = np.ndarray(
            shape=(1,224,224,3),
            dtype=np.float32
        )

        data[0] = normalized

        # =================================================
        # PREDICT
        # =================================================

        try:

            prediction = model.predict(
                data,
                verbose=0
            )

            index = np.argmax(prediction)

            confidence = float(
                prediction[0][index]
            ) * 100

            raw_class_name = class_names[index]

            parts = raw_class_name.split(" ",1)

            label_key = (
                parts[1].strip().lower()
                if len(parts) > 1
                else raw_class_name.lower()
            )

            if confidence < 70:

                label_key = "unknown"

        except Exception as e:

            print(e)

            label_key = "unknown"

            confidence = 0

        # =================================================
        # PRICE
        # =================================================

        if label_key in PRICE_MAP:

            dish_name = PRICE_MAP[label_key]["name"]

            price = PRICE_MAP[label_key]["price"]

        else:

            dish_name = "Khong xac dinh"

            price = 0

        total_bill += price

        pdf_items.append({

            "slot": f"Ngan {i+1}",

            "name": dish_name,

            "confidence": confidence,

            "price": price
        })

        # =================================================
        # HTML
        # =================================================

        html += f"""

        <div style='
            background:#1e293b;
            padding:15px;
            border-radius:12px;
            margin-bottom:10px;
        '>

        <b>{dish_name}</b>

        <br>

        AI: {confidence:.1f}%

        <br>

        Gia: {price:,.0f} VND

        </div>
        """

        # =================================================
        # SHOW
        # =================================================

        axes[i].imshow(crop)

        axes[i].set_title(
            dish_name
        )

        axes[i].axis("off")

    # =====================================================
    # TOTAL
    # =====================================================

    html += f"""

    <hr>

    <h2 style='color:#a855f7;'>

    Tong: {total_bill:,.0f} VND

    </h2>

    </div>
    """

    # =====================================================
    # SAVE CROPS
    # =====================================================

    plt.tight_layout()

    plt.savefig(
        "temp_crops.png",
        bbox_inches='tight',
        dpi=150
    )

    plt.close(fig)

    # =====================================================
    # FIX IMAGE FOR GRADIO
    # =====================================================

    result_crops_img = cv2.imread(
        "temp_crops.png"
    )

    result_crops_img = cv2.cvtColor(
        result_crops_img,
        cv2.COLOR_BGR2RGB
    )

    # =====================================================
    # QR
    # =====================================================

    if os.path.exists(QR_PATH):

        qr_image = cv2.imread(QR_PATH)

        qr_image = cv2.cvtColor(
            qr_image,
            cv2.COLOR_BGR2RGB
        )

    else:

        qr_image = np.ones(
            (300,300,3),
            dtype=np.uint8
        ) * 255

    # =====================================================
    # PDF
    # =====================================================

    pdf_file_path = generate_pdf_invoice(
        pdf_items,
        total_bill
    )

    return (

        result_crops_img,

        html,

        qr_image,

        pdf_file_path
    )

# =========================================================
# CSS
# =========================================================

custom_css = """

body{

background:
linear-gradient(
135deg,
#eef2ff,
#dbeafe,
#f5f3ff
);

}

"""

# =========================================================
# UI
# =========================================================

with gr.Blocks(
    title="AI Smart Canteen",
    css=custom_css
) as demo:

    gr.Markdown(
        "# 🍽 AI SMART CANTEEN"
    )

    with gr.Row():

        with gr.Column():

            camera = gr.Image(

                sources=[
                    "upload",
                    "webcam"
                ],

                type="numpy",

                height=450
            )

            tray_layout = gr.Radio(

                [
                    "3 TOP - 2 BOTTOM",
                    "2 TOP - 3 BOTTOM"
                ],

                value="3 TOP - 2 BOTTOM",

                label="Kieu khay"
            )

            btn = gr.Button(
                "🔍 QUET KHAY COM"
            )

        with gr.Column():

            output_crops = gr.Image(
                label="AI Detection"
            )

            output_qr = gr.Image(
                label="QR"
            )

        with gr.Column():

            output_html = gr.HTML()

            output_pdf = gr.File(
                label="Tai PDF"
            )

    btn.click(

        fn=process_and_bill,

        inputs=[
            camera,
            tray_layout
        ],

        outputs=[

            output_crops,

            output_html,

            output_qr,

            output_pdf
        ]
    )

# =========================================================
# RUN
# =========================================================

demo.launch(
    debug=True,
    share=True,
    inline=True
)
```
